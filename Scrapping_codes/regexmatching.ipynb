```python
# Create a GCS client
client = storage.Client()
```

```python
# Compile a regular expression for matching arXiv IDs
# This pattern will match 'arXiv:' followed by numbers that include a period.
arxiv_id_pattern = re.compile(r'(arXiv:\d{4}\.\d{4,5})(v\d+)?')
```

```python
# Define a function to extract arXiv IDs from the PDF text
def extract_arxiv_ids_from_pdf(pdf_bytes):
    pdf_stream = io.BytesIO(pdf_bytes)
    reader = PyPDF2.PdfReader(pdf_stream)

    # Extract text and arXiv IDs from each page
    arxiv_ids = []
    for page in reader.pages:
        text = page.extract_text()
        if text:  # If text extraction is successful
            found_arxiv_ids = arxiv_id_pattern.findall(text)
            arxiv_ids.extend(found_arxiv_ids)

    return arxiv_ids
```

```python
# Get the bucket object
bucket = client.bucket(bucket_name)
```

```python
# List and process PDF blobs in the specified folder within the bucket
res = []
for blob in client.list_blobs(bucket, prefix=folder_prefix):
    # Ensure that you are only processing PDF files
    if not blob.name.lower().endswith('.pdf'):
        continue

    print(f"Processing {blob.name}...")

    # Download the PDF blob as bytes
    pdf_bytes = blob.download_as_bytes()

    # Extract arXiv IDs from the PDF
    arxiv_ids = extract_arxiv_ids_from_pdf(pdf_bytes)

    # Do something with the extracted arXiv IDs (print, store, etc.)
    print(f"Found arXiv IDs in {blob.name}: {arxiv_ids}")
    for id in arxiv_ids:
      res.append([blob.name, id[0]])
```

```python
res
```

```python
import pandas as pd

# Assuming 'res' is a list of tuples like [('arxiv_pdfs/2403.07865v2.pdf', 'arXiv:2403.07865'), ...]

# Convert 'res' into a DataFrame
df = pd.DataFrame(res, columns=['paper', 'refers'])

# Aggregate references into a list for each paper
df_aggregated = df.groupby('paper')['refers'].apply(list).reset_index()

# Convert the list of references into a string representation
df_aggregated['refers'] = df_aggregated['refers'].apply(lambda x: ', '.join(x))

# Save to CSV
df_aggregated.to_csv('paper_references.csv', index=False)
```
