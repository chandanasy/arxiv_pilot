```python
import requests
import arxiv
import os
import time

# Function to download a PDF from a given URL
def download_pdf(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f'Downloaded {filename}')
    else:
        print(f'Error downloading {filename}: Status code {response.status_code}')

# Search parameters
query = 'quantum computing'  # Replace with your topic of interest
max_results = 5  # Number of PDFs you want to download

# Perform search
search = arxiv.Search(
    query = query,
    max_results = max_results,
    sort_by = arxiv.SortCriterion.SubmittedDate
)

# Directory to save PDFs
save_dir = 'arxiv_pdfs'
os.makedirs(save_dir, exist_ok=True)

# Download PDFs
for result in search.get():
    # Create a valid filename
    title = result.title.replace(' ', '_').replace('/', '_')
    filename = os.path.join(save_dir, f'{title}.pdf')

    # URL for the PDF
    pdf_url = result.pdf_url

    # Download the PDF
    download_pdf(pdf_url, filename)

    # Sleep to be respectful of arXiv's servers
    time.sleep(5)  # Sleep for 5 seconds between downloads

print('Download complete.')
```
